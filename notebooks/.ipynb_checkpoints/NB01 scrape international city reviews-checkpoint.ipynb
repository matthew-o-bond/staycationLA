```python
# function to check if the button is on the page, to avoid miss-click problem
def check_exists_by_xpath(xpath):
    try:
        driver.find_element_by_xpath(xpath)
    except NoSuchElementException:
        return False
    return True
```

```python
# function to count number of webpages per international city in tripadvisor
#multiple if/else statements account for source code differences on different location pages
def city_scraper(driver):
    pageCounter = 0
    #maxPageCount = 52
    html_soup = BeautifulSoup(driver.page_source, 'html.parser')

    #pull out list of page numbers
    post_content = html_soup.find_all('a', class_ = 'paging taLnk')

    #find last element of text in list of page numbers
    last_div = None
    for last_div in post_content:pass
    if last_div:
        content = last_div.getText()

    maxPageCount = int(content)+1
    
    while (pageCounter < maxPageCount):

        container = driver.find_elements_by_xpath("//div[@class='  reviewSelector ']")
        num_page_items = len(container);

        for j in range(num_page_items):
            # to save the rating
            string = container[j].find_element_by_xpath(".//span[contains(@class, 'ui_bubble_rating bubble_')]").get_attribute("class")
            data = string.split("_")
            # to save in a csv file readable the star and the review [Ex: 50,"I love this place"]
            csvWriter.writerow([data[3], container[j].find_element_by_xpath(".//div[@class='quote']").text.replace("\n", ""), container[j].find_element_by_xpath(".//div[@class='entry']").text.replace("\n", "")])

        # to change the page
        if (check_exists_by_xpath('//a[@class="sbx_close"]')==True):
            driver.find_element_by_xpath('//a[@class="sbx_close"]').click()
            time.sleep(2)
            driver.find_element_by_xpath('//a[@class="guiArw sprite-pageNext "]').click()
            time.sleep(5)
            pageCounter +=1
            print(pageCounter)        
        elif (check_exists_by_xpath('//a[@class="guiArw sprite-pageNext "]')==True):
            # to click next
            driver.find_element_by_xpath('//a[@class="guiArw sprite-pageNext "]').click()
            time.sleep(5)
            pageCounter +=1
            print(pageCounter)
            #if popup banner disrupts

        else:
            #to get reviews on last page
            csvFile.flush() #makes sure all scraped content is in csv file
            pageCounter +=1
            print(pageCounter)
            pageCounter +=1
            print("end")
            
    csvFile.flush() #makes sure all scraped content is in csv file     
    driver.close()     
```

```python
translator = Translator()
trans = lambda x: translator.translate(x)
```

```python
def positive_review_string(reviews):
    #set rating to be 4-5 stars only
    reviews = reviews[reviews['stars'] > 35]
    reviews["all_text"] = reviews["title"] + " " + reviews["text"]
    reviews['all_text2']=reviews['all_text'].apply(trans) #translate
    
    reviews.drop(['title', 'text', 'all_text'], axis=1)
    reviews['all_text2'] = reviews['all_text2'].astype(str)

    #make into list
    reviews_texts = reviews['all_text2'].tolist()
    #type(reviews_texts_pos)

    #make into string
    # initialize an empty string 
    str1 = " " 
    # return string   
    reviews=(str1.join(reviews_texts)) 
    return(reviews)
```

```python
import string
def clean_text(text):
    '''Make text lowercase, remove text in square brackets, remove punctuation and remove words containing numbers.'''
    text = text.replace('Translated(src=en, dest=en, text=“', ' ')
    text = text.replace('pronunciation=[[]], extra_data=""{', ' ').replace('translat...")', ' ').replace('translat', ' ')
    #text = text.replace('en, ', ' ')
    text = text.lower()
    text = re.sub(r'\[.*?\]', '', text)
    text = re.sub(r'[%s]' % re.escape(string.punctuation), ' ', text)
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = re.sub(r'\w*\d\w*', '', text)
    text = text.replace('“', '').replace('”', '').replace('"', '') 
    text = text.replace(',', ' ').replace('’', ' ').replace(' i ', ' ').replace('  ', ' ')
    text = text.replace('「', ' ').replace('」', ' ').replace('：', ' ').replace('–', ' ')
    text = text.replace('。', ' ').replace('、', ' ').replace(',', ' ').replace(' ̇ ', ' ')
    text = text.replace('pronunciation', ' ').replace('extra dat', ' ').replace('a text', ' ').replace(' a ', ' ')
    text = text.replace('・', ' ').replace('，', ' ').replace('…', ' ').replace('～', ' ').replace('？', ' ')
    text = text.replace('（', ' ').replace('）', ' ').replace('—', ' ').replace('°', ' ')
    text = text.replace('『', ' ').replace('』', ' ').replace('♪', ' ').replace('》', ' ').replace('《', ' ')    
    return text

cities['clean'] = (cities.review_text.apply(lambda x: clean_text(x)))
pd.options.display.max_colwidth = 50
cities.tail(10)
```

```python
import nltk.corpus
nltk.download('stopwords')
from nltk.corpus import stopwords
stop = stopwords.words('english')
cities['clean2'] = cities['clean'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
cities.head()
```

```python
#nlp = spacy.load('en')
nlp = spacy.load("en_core_web_sm")
#nlp = spacy.load(‘en_core_web_lg’)
def lemmatizer(text):        
    sent = []
    doc = nlp(text)
    for word in doc:
        sent.append(word.lemma_)
    return " ".join(sent)


cities["lemmatize"] =  cities.apply(lambda x: lemmatizer(x['clean2']), axis=1)
cities['lemmatize_clean'] = cities['lemmatize'].str.replace('-PRON-', '')
```
